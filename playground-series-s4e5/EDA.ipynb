```python
import pandas as pd
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load data
def load_data():
    train = pd.read_csv('../playground-series-s4e5/train.csv')
    test = pd.read_csv('../playground-series-s4e5/test.csv')
    return train, test

# define dataframe
train, test = load_data()
```

```python
# check the first 5 rows of the dataframe
train.head()
```

```python
# Check for missing values
def missing_values(df):
    return df.isnull().sum()

# Check for missing values
print(missing_values(train))
print(missing_values(test))
```

```python
# Check for duplicates
def duplicates(df):
    return df.duplicated().sum()

# Check for duplicates
print(duplicates(train))
print(duplicates(test))
```

```python
# check for the range of values in the dataframe
def range_values(df):
    return df.describe()

print(range_values(train))
print(range_values(test))
```

```python
# draw a correlation matrix heatmap
def correlation_matrix(df):
    corr = df.corr()
    sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=2)
    plt.show()

correlation_matrix(train)
```

```python
# check for the distribution of the target variable
def target_distribution(df):
    sns.countplot(df['FloodProbability'])
    plt.show()

target_distribution(train)
```
